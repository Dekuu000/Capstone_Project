# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import matplotlib as mpl
import matplotlib.pyplot as plt

QUERY = """SELECT 
ConvertedComp 
FROM master"""

df = pd.read_sql_query(QUERY, conn)

df['ConvertedComp'].plot(kind='hist', figsize=(8,5))
plt.title('Histogram of ConvertedComp')
plt.xlabel('Annual USD')
plt.ylabel('Number of Respondents')
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """SELECT Age 
FROM master"""

df = pd.read_sql_query(QUERY,conn)
df['Age'].plot(kind='box', figsize=(8,6))
plt.title('Boxplot of Age')
plt.ylabel('Age Interval')
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

QUERY = """SELECT Age, 
WorkWeekHrs FROM master"""

df = pd.read_sql_query(QUERY, conn)
df.plot(kind='scatter', x='Age', y='WorkWeekHrs', figsize=(10, 6), color='lightblue')
plt.title('Scatter Plot of Age and WorkWeekHrs')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """SELECT WorkWeekHrs, 
CodeRevHrs, Age FROM master"""

df_bubble = pd.read_sql_query(QUERY, conn)
norm_Age = (df_bubble['Age'] - df_bubble['Age'].min()) / (df_bubble['Age'].max() - df_bubble['Age'].min())
df_bubble.plot(kind='scatter',
               x='WorkWeekHrs',
               y='CodeRevHrs',
               figsize=(10,7),
               alpha=0.5,
               color='green',
               s=norm_Age * 1000 + 10)
plt.title('Relation Between WorkWeekHrs and CodeRevHrs of Respondents')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.legend(['Age'], loc = 'upper right', fontsize='x-large')

plt.show()
```

# Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY = """
SELECT * 
FROM DatabaseDesireNextYear
"""

df_db = pd.read_sql_query(QUERY, conn)
df_groupby = df_db.groupby('DatabaseDesireNextYear', axis=0).count()
df_groupby.sort_values(["Respondent"], ascending=False, inplace=True)
df_pie = df_groupby.head()
df_pie['Respondent'].plot(kind='pie',
                    figsize=(5, 6),
                    autopct='%1.1f%%',
                    startangle=90,
                    shadow=True)
plt.title('Top 5 Databases that Respondents wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT CodeRevHrs, WorkWeekHrs, Age
FROM master
"""
df_stacked = pd.read_sql_query(QUERY,conn)
df_stacked = df_stacked.loc[(df_stacked['Age'] >= 30) & (df_stacked['Age'] <= 35)].reset_index().drop(columns=['index'])

df_stacked.groupby(['CodeRevHrs','Age'])['WorkWeekHrs'].median().sort_values(ascending=False).unstack().plot(kind='bar',
                                                                                                             stacked=True, 
                                                                                                             figsize=(25, 10))
plt.title('Stacked Chart of Median WorkWeekHrs and CodeRevHrs For The Age Group 30 to 35',
          loc='center',pad=30, fontsize=30)
plt.xlabel('CodeRevHrs', fontsize = 20)
plt.ylabel('WorkWeekHrs', fontsize = 20)
plt.legend(bbox_to_anchor=(1, 1), shadow=True, ncol=1, title='Age')
plt.ylim(ymin=0)
labels, locations = plt.yticks()
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY = """SELECT ConvertedComp, Age 
FROM master"""

df_line = pd.read_sql_query(QUERY, conn)
df_line = df_line.loc[(df_line['Age'] >= 45 ) & (df_line['Age'] <= 60)].reset_index().drop(columns=['index'])
df_line.dropna(inplace=True)
df_line = df_line.groupby('Age', axis=0).median()
df_line

df_line.plot(kind="line", color='orange')
plt.title("Average Compensation Based on Age Groups")
plt.ylabel("Converted Compensation (Annual USD)")
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY = """SELECT MainBranch FROM master"""

df_hbar = pd.read_sql_query(QUERY, conn)
df_hbar = df_hbar.groupby('MainBranch')['MainBranch'].count()
df_hbar.plot(kind='barh',
             figsize=(10, 6))
plt.show()
```
